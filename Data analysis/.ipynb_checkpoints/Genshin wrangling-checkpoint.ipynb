```python
rev["start_date"] = pd.to_datetime(rev["start_date"])
rev["end_date"] = pd.to_datetime(rev["end_date"])
rev['revenue'] = rev['revenue'].str.replace(',', '').astype(int)
# rev['avg_revenue'] = rev['avg_revenue'].str.replace(',', '').astype(float)
```

```python
rev.to_csv('genshinRev.csv', index=False)
```

```python
columns_to_clean = ['spsecial_0', 'spsecial_1', 'spsecial_2', 'spsecial_3', 'spsecial_4', 'spsecial_5', 'spsecial_6']

for column in columns_to_clean:
    if column in df:
        df[column] = df[column].astype(str).str.replace('%', '', regex=False).astype(float)
```

```python
df.to_csv('genshinChar.csv', index=False)
```

```python
fiveStar = df.loc[df['rarity'] == 5]
indices_to_delete = [2,14,24,40,45,57,58]
fiveStar = fiveStar.drop(indices_to_delete)
fiveStar.to_csv("fiveStarChar.csv")
```
